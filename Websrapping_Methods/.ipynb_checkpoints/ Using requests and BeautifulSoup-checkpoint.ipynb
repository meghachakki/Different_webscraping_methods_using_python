```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL to scrape
url = 'https://www.w3schools.com/'

# Send GET request
response = requests.get(url)

# Parse HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all tutorial links
tutorials = soup.select('div.w3-bar-block a.w3-bar-item')
data=[]
# Extract and print titles and URLs
for tutorial in tutorials:
    title = tutorial.text.strip()
    link = tutorial['href']
    full_link = f"https://www.w3schools.com{link}"
    data.append((title, full_link))

    print(f'Title: {title}\nURL: {full_link}\n')

```

```python
print(data)
```

```python
df = pd.DataFrame(data, columns=['Title', 'URL'])

# Print the DataFrame
print(df.to_string())

# Optionally, save the DataFrame to a CSV file
df.to_csv('w3schools_tutorials1.csv', index=False)
```
