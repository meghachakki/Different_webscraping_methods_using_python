```python
from lxml import html
import requests
import pandas as pd
```

```python
# URL to scrape
url = 'https://www.w3schools.com/'

# Send GET request
response = requests.get(url)

# Parse HTML content
tree = html.fromstring(response.content)

# Find all tutorial links (adjust XPath based on actual HTML structure)
tutorials = tree.xpath('//a[contains(@class, "w3-bar-item") and contains(@class, "w3-button")]')

# Use a set to store unique links
unique_tutorials = set()

# Filter and store unique tutorial links
for tutorial in tutorials:
    text = tutorial.text_content().strip()
    href = tutorial.get('href')
    if href and text:
        unique_tutorials.add((text, href))

# Print the unique tutorial links
for text, href in sorted(unique_tutorials):
    print(f"{text}: {href}")
```

```python
df = pd.DataFrame(list(unique_tutorials), columns=['Tutorial', 'Link'])

# Print the DataFrame
print(df.to_string())

# Optionally, save the DataFrame to a CSV file
df.to_csv('w3schools_tutorials.csv', index=False)
```
